Question 1: randomly generate x and y using random module and do linear regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
x=np.random.randint(100,1000,15).reshape((-1,1))
y=np.random.randint(100,1000,15)
#same data in both x and y
plt.scatter(x,y)
plt.title(" linear regression")
plt.xlabel("X- AXIS")
plt.ylabel("Y- AXIS")
plt.show()

#preparing for the process
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
#training the model
SLR.fit(x,y)
#y^
prediction=SLR.predict(x)

from sklearn.metrics import mean_squared_error as mse
mse=(y,prediction)
plt.plot(x,prediction,color='b',marker='*')

print("MSE: ",mse)
print("intercept: ",SLR.intercept_)
print("slope: ",SLR.coef_)
```

Question 2:
Read an excel file to fetch x and y values and repeat linear regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df=pd.read_excel('Historicalinvesttemp.xlsx')
print(df)

#Excel file has columns named Year and stocks
x=df['Year'].values.reshape(-1,1)
y=df['Stocks'].values

#Linear Regression Model

SLR=LinearRegression()

#training the model
SLR.fit(x,y)
predicted_stock_values=SLR.predict(x)

print(predicted_stock_values)

plt.scatter(x,y,color='red',label='Actual Data')
plt.plot(x,predicted_stock_values,color='blue',label='Regression Line')
plt.title("Linear Regression:Year vs stock Value")
plt.xlabel("Year")
plt.ylabel("Stocks")
plt.legend()
plt.show()

#Calculating Mean squared error

mse_value=mean_squared_error(x,predicted_stock_values)


print("MSE:", mse_value)
print("Intercept:", SLR.intercept_)
print("Slope:", SLR.coef_)






```
