```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


data1=pd.read_csv('/content/sample_data/california_housing_train.csv')
print(data1)
```

```python
x=data1[['housing_median_age','total_bedrooms','total_rooms']]
y=data1['population']

print(x)
print(y)
```

```python


plt.plot(x,y,color='g',marker="*")
plt.xlabel("independent")
plt.ylabel("dependent")
plt.show()
```

```python
SLR=LinearRegression()
SLR.fit(x,y)
pred_population=SLR.predict(x)

mse=mean_squared_error(y,pred_population)
plt.plot(x,y,color='b',marker="*")
plt.plot(x,pred_population,color='r',marker='*')
plt.title("linear-multi")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

print("mse:",mse)
print("intercept",SLR.intercept_)
print("slope:",SLR.coef_)
```
